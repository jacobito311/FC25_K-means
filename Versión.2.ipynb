```python
import numpy as np  # Cálculos numéricos
import pandas as pd # Manipulación de datos
import matplotlib.pyplot as plt # Visualización de gráficos
import seaborn as sns # Gráficos atractivos
from sklearn.cluster import KMeans  # Agrupación de datos
from sklearn.decomposition import PCA # Reducción de dimensionalidad
```

```python
url = "https://docs.google.com/spreadsheets/d/e/2PACX-1vSLo2NoEd47ZPZnEeDQBO95NHxvMsUi7YG4nH9DEiDfMJIV2q5Qk0fPQX5J10ClRB7ixPrCbfkTHcTh/pub?output=csv"  # URL del archivo CSV
data = pd.read_csv(url)  # Carga los datos desde la URL en un DataFrame de pandas
```

```python
# Definir las columnas de interés
columnas_objetivo = [  # Lista de columnas que se conservarán
    'Name', 'OVR', 'PAC', 'SHO', 'PAS', 'DRI', 'DEF', 'PHY',
    'Acceleration', 'Sprint Speed', 'Positioning', 'Finishing',
    'Shot Power', 'Long Shots', 'Volleys', 'Penalties',
    'Vision', 'Crossing', 'Free Kick Accuracy', 'Short Passing',
    'Long Passing', 'Curve', 'Dribbling', 'Agility',
    'Balance', 'Reactions', 'Ball Control', 'Composure',
    'Interceptions', 'Heading Accuracy', 'Def Awareness',
    'Standing Tackle', 'Sliding Tackle', 'Jumping',
    'Stamina', 'Strength', 'Aggression', 'Weak foot',
    'Skill moves', 'Age', 'GK Diving', 'GK Handling',
    'GK Kicking', 'GK Positioning', 'GK Reflexes',
    'Preferred foot', 'Nation', 'Position'
]

columnas = data.columns  # Obtiene todas las columnas del DataFrame 'data'

# Iterar sobre las columnas del DataFrame
for col in columnas:
    if col not in columnas_objetivo:  # Si la columna no está en la lista de interés
        data.drop(col, axis=1, inplace=True)  # Elimina la columna del DataFrame
    else:
        data[col] = data[col]  # Re-asigna la columna (sin cambios)
```

```python
# Imputar datos faltantes para las categorías de GK cambiando NaN por 0
gk_columns = [  # Lista de columnas relacionadas con el portero (GK)
    'GK Diving', 'GK Handling', 'GK Kicking',
    'GK Positioning', 'GK Reflexes'
]

# Rellena los valores NaN en las columnas de GK con 0
data[gk_columns] = data[gk_columns].fillna(0)
```

```python
### Ahora extraeremos los datos
X = data.iloc[:, 1:-1].values  # Extrae todos los valores del DataFrame 'data' excepto la primera columna y la última (variable objetivo)
```

```python
### Codificar el pie preferido como variable numérica
from sklearn.preprocessing import LabelEncoder  # Importa LabelEncoder para convertir variables categóricas a numéricas

# Crear el codificador
le = LabelEncoder()  # Instancia del codificador de etiquetas

# Codificar la penúltima columna (Preferred foot)
X[:, -2] = le.fit_transform(X[:, -2])  # Codifica 'Preferred foot' y reemplaza los valores originales en el array X

# Codificar la última columna (Nation)
X[:, -1] = le.fit_transform(X[:, -1])  # Codifica 'Nation' y reemplaza los valores originales en el array X
```

```python
df = pd.DataFrame(X)  # Crea un DataFrame a partir del array X

# Iterar sobre cada columna del DataFrame
for col in df.columns:
    # Convertir a numérico, forzando errores a NaN
    df[col] = pd.to_numeric(df[col], errors='coerce')  # Convierte cada columna a tipo numérico, reemplazando errores con NaN
```

```python
X = df.values  # Extrae los valores del DataFrame 'df' y los convierte en un array de NumPy
```

```python
from sklearn.impute import SimpleImputer  # Importa SimpleImputer para manejar datos faltantes

# Crear un objeto imputer que reemplaza NaN con la mediana de la columna
imputer = SimpleImputer(missing_values=np.nan, strategy='median')

imputer.fit(X)  # Ajusta el imputer a los datos X para calcular la mediana

X = imputer.transform(X)  # Aplica la transformación, reemplazando los NaN en X con la mediana
```

```python
X_ = X[:,:-1]  #### Eligiendo todos los datos, excepto la nacionalidad
```

```python
from sklearn.cluster import KMeans  # Importa KMeans para realizar el clustering

# Inicializo donde almacenaré el WCSS (Within-Cluster Sum of Squares)
WCSS = []  # Lista para almacenar los valores de WCSS para diferentes valores de k

# Calcular WCSS para valores de k de 1 a 10
for k in range(1, 11):
    kpp = KMeans(n_clusters=k,  # Inicializa KMeans con el número de clústeres k
                 init='k-means++',  # Método de inicialización
                 random_state=42,   # Fija la semilla para reproducibilidad
                 n_init='auto')     # Usa el valor predeterminado para el número de inicializaciones

    kpp.fit(X_)  # Ajusta el modelo a los datos X_
    WCSS.append(kpp.inertia_)  # Almacena el WCSS para este valor de k

# Graficar el resultado
plt.figure()  # Crea una nueva figura
plt.plot(range(1, 11), WCSS, 'o--', color='salmon')  # Grafica los valores de WCSS en función de k
plt.title('Método del codo KMeans ++')  # Título del gráfico
plt.xlabel('k')  # Etiqueta del eje X
plt.ylabel('WCSS')  # Etiqueta del eje Y
plt.grid(True)  # Activa la cuadrícula
plt.grid(linestyle='--')  # Define el estilo de las líneas de la cuadrícula como discontinuas
plt.show()  # Muestra el gráfico
```

```python
from sklearn.cluster import KMeans  # Importa KMeans para realizar el clustering

# Inicializa el modelo KMeans con 4 clústeres y una semilla para reproducibilidad
kmeans_mas = KMeans(n_clusters=4, random_state=42, init='k-means++')

# Ajusta el modelo a los datos X_ y almacena las etiquetas de los clústeres
clusters_kmeans_mas = kmeans_mas.fit_predict(X_)
```

```python
### Ver centroides
centroides_kmeans_mas = kmeans_mas.cluster_centers_  # Obtiene las coordenadas de los centroides de los clústeres
centroides_kmeans_mas  # Muestra los centroides en la salida
```

```python
from sklearn.decomposition import PCA  # Importa PCA para la reducción de dimensionalidad

X_ = X[:, :-1]  # Elige todos los datos de X, excepto la última columna (nacionalidad)

# Crear un objeto PCA y ajustar los datos
pca = PCA(n_components=3)  # Queremos reducir a las tres primeras componentes principales
X_pca = pca.fit_transform(X_)  # Ajusta PCA a los datos y transforma X_ en las componentes principales
```

```python
X_pca = pca.fit_transform(X_)  # Ajusta PCA a los datos X_ y transforma los datos a las componentes principales
```

```python
def graficar_clusters_PCA2D(clusters, metodo, centroids):
    # Crear un gráfico de dispersión 2D usando las dos primeras componentes
    plt.figure(figsize=(20, 10))  # Establece el tamaño de la figura
    sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1],  # Dibuja los puntos de dispersión
                    hue=clusters, palette='viridis',  # Colorea por clúster
                    alpha=0.6, s=100)  # Establece la transparencia y el tamaño de los puntos

    # Dibuja los centroides en el gráfico
    plt.scatter(centroids[:, 0], centroids[:, 1], c='black', s=100, label='centroides')
    plt.title(f'Visualización 2D de la Clusterización (PCA) {metodo}', fontsize=16)  # Título del gráfico
    plt.xlabel('Componente Principal 1')  # Etiqueta del eje X
    plt.ylabel('Componente Principal 2')  # Etiqueta del eje Y
    plt.grid(True)  # Activa la cuadrícula
    plt.grid(linestyle='-', alpha=0.5)  # Estilo de la cuadrícula
    plt.legend(title='Cluster')  # Muestra la leyenda
    plt.show()  # Muestra el gráfico
```

```python
from mpl_toolkits.mplot3d import Axes3D  # Importa las herramientas para graficar en 3D

def graficar_clusters_PCA3D(clusters, metodo):
    # Crear el gráfico 3D
    fig = plt.figure(figsize=(15, 6))  # Establece el tamaño de la figura
    ax = fig.add_subplot(111, projection='3d')  # Añade un subplot en 3D

    # Graficar los puntos con borde blanco
    scatter = ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2],  # Coordenadas en 3D
                         c=clusters, cmap='viridis', s=100,  # Colorea por clúster
                         edgecolor='white', alpha=0.6)  # Borde blanco y transparencia

    # Etiquetas de los ejes y título
    ax.set_title(f'Visualización 3D de la Clusterización (PCA) {metodo}', fontsize=16)  # Título del gráfico
    ax.set_xlabel('Componente Principal 1')  # Etiqueta del eje X
    ax.set_ylabel('Componente Principal 2')  # Etiqueta del eje Y
    ax.set_zlabel('Componente Principal 3')  # Etiqueta del eje Z

    # Añadir la leyenda
    legend1 = ax.legend(*scatter.legend_elements(), title="Clusters")  # Crea la leyenda
    ax.add_artist(legend1)  # Añade la leyenda al gráfico

    # Mostrar el gráfico
    plt.show()  # Muestra el gráfico
```

```python
centroides_kmeans_mas_pca = pca.transform(centroides_kmeans_mas)  # Transforma los centroides de KMeans a las componentes principales usando PCA
```

```python
graficar_clusters_PCA2D(clusters_kmeans_mas, 'K-means ++', centroides_kmeans_mas_pca)  # Llama a la función para graficar los clusters en 2D usando PCA
```

```python
graficar_clusters_PCA3D(clusters_kmeans_mas, 'K-means ++')  # Llama a la función para graficar los clusters en 3D usando PCA
```
