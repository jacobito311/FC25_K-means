***Veamos cuáles son las 10 ligas con el mayor número de jugadores:***



```python
# value_counts() obtiene la cuenta de ocurrencias de cada valor en la columna "League" del DataFrame 'data'.
# head(10) selecciona los primeros 10 conteos más altos.
ligas = original_data["League"].value_counts().head(10)

# Gráfica
plt.figure(figsize=(8, 6))
palette = sns.color_palette("coolwarm", len(ligas))  # Usamos una paleta de colores "coolwarm"
ligas.plot(kind='barh', color=palette)

# Añadimos los valores a las barras
for index, value in enumerate(ligas):
    plt.text(value + 1, index, str(value), va='center', fontsize=10)

# Mejoramos el diseño
plt.title('Top 10 Ligas con Mayor Cantidad de Jugadores', fontsize=16, fontweight='bold')
plt.xlabel('Cantidad de Jugadores', fontsize=12)
plt.ylabel('Ligas', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

# Ajustar márgenes
plt.subplots_adjust(left=0.3, right=0.95)

# Fondo blanco y sin bordes alrededor del gráfico
sns.despine(left=True, bottom=True)

plt.show()
```

***Análisis por nacionalidad:***


```python
# 1. Contar el número de jugadores por nacionalidad
nacionalidad_counts = data['Nation'].value_counts().reset_index()
nacionalidad_counts.columns = ['Nation', 'Count']

# 2. Calcular el OVR promedio por nacionalidad
nacionalidad_avg_ovr = data.groupby('Nation')['OVR'].mean().reset_index()
nacionalidad_avg_ovr.columns = ['Nation', 'Average OVR']

# 3. Combinar ambos DataFrames
nacionalidad_analysis = pd.merge(nacionalidad_counts, nacionalidad_avg_ovr, on='Nation')

# 4. Visualizar la distribución de jugadores por nacionalidad (Top 15)
plt.figure(figsize=(6, 5))
top_nacionalidades = nacionalidad_analysis.nlargest(15, 'Count')
sns.barplot(x='Count', y='Nation', data=top_nacionalidades, palette='viridis', hue = 'Nation')
plt.title('Top 15 Nacionalidades por Número de Jugadores', fontsize=16)
plt.xlabel('Número de Jugadores')
plt.ylabel('Nacionalidad')
plt.show()

# 5. Visualizar el OVR promedio por nacionalidad (Top 15)
# plt.figure(figsize=(6, 5))
# top_avg_ovr = nacionalidad_analysis.nlargest(15, 'Average OVR')
# sns.barplot(x='Average OVR', y='Nation', data=top_avg_ovr, palette='plasma', hue = 'Nation')
# plt.title('Top 15 Nacionalidades por Calificación General Promedio (OVR)', fontsize=16)
# plt.xlabel('Calificación General Promedio (OVR)')
# plt.ylabel('Nacionalidad')
# plt.show()

# 6. Opcional: guardar el análisis en un archivo CSV
nacionalidad_analysis.to_csv('nacionalidad_analysis.csv', index=False)
```

```python
# 5. Visualizar el OVR promedio por nacionalidad (Top 15)
plt.figure(figsize=(6, 5))
top_avg_ovr = nacionalidad_analysis.nlargest(15, 'Average OVR')
sns.barplot(x='Average OVR', y='Nation', data=top_avg_ovr, palette='plasma', hue = 'Nation')
plt.title('Top 15 Nacionalidades por Calificación General Promedio (OVR)', fontsize=16)
plt.xlabel('Calificación General Promedio (OVR)')
plt.ylabel('Nacionalidad')
plt.show()
```

***Distribución del Overral (OVR):***

```python
##Creamos un histograma:

# Crea un arreglo de valores desde 40 hasta 99 (con paso de 1 en 1), que servirá como los límites de los intervalos (bins).
bins_ = np.arange(40, 100, 1)



plt.figure(figsize=(6, 4))
# Crea un histograma de la columna "OVR" del DataFrame 'data' utilizando Seaborn.
sns.histplot(
    x="OVR",             # Especifica la columna del DataFrame que se quiere graficar.
    data=data,          # Indica el DataFrame que contiene los datos.
    bins=bins_,         # Establece los límites de los intervalos (bins) para el histograma.
    kde=True,           # Agrega una curva de densidad (Kernel Density Estimate) al histograma.
    color='g'           # Define el color de las barras del histograma (verde).
)
plt.xlabel('Overall Rating')
plt.ylabel('Frecuencia')
plt.title('Distribución del Overall')
plt.grid(True)
plt.grid('--', alpha = 0.6)
plt.show()
```

***OVR por posición:***

```python
# Agrupa los datos por la columna 'Position' y calcula la mediana de la columna 'OVR' para cada grupo.
ovr_posiciones = data.groupby('Position')['OVR'].median()

# Ordena los valores resultantes de menor a mayor.
ovr_posiciones = ovr_posiciones.sort_values()

ovr_posiciones
```

```python
import warnings
warnings.filterwarnings('ignore')
## Diagrama de cajas (Boxplot)
plt.figure(figsize=(6, 4))
sns.boxplot(x = 'Position', # Nombre de la variable en el DataFrame
            y = 'OVR',      ## Nombre de la variable dependiente
            data = data,    ## Conjunto de datos
            order = ovr_posiciones.index, ## variable creada anteriormente para los índices
                                          #Usamos '.index' para obtener las etiquetas de la serie 'ovr_posiciones'.
            palette = 'Accent')
plt.xlabel('Posición')
plt.ylabel('Overall')
plt.title('Overall por posición')
plt.show()
```

***Atributos por posición:***

```python
import warnings
warnings.filterwarnings('ignore')

attributes = ['PAC', 'SHO', 'DRI', 'PHY', 'DEF', 'PAS']
# Set up the figure and axes for subplots
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(20, 20), sharey=True,)
fig.suptitle('Player attributes by position', fontsize=16)

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Plot the distribution of each attribute in a separate subplot
for i, attr in enumerate(attributes):
    order = data.groupby('Position')[attr].median().sort_values().index

    # Plot the boxplot with ordered categories
    sns.boxplot(x='Position', y=attr, data=data, ax=axes[i], order=order, palette = 'PiYG')
    axes[i].set_title(f'{attr} rating by position')
    axes[i].set_ylabel(f'{attr}')
    axes[i].set_xlabel('Position')
# Adjust the layout to prevent overlap
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

***¿La edad afecta el OVR?:***

```python
# Crea una figura y un conjunto de ejes para el gráfico con un tamaño de 6x5 pulgadas.
fig, axes = plt.subplots(figsize=(6, 5))

# Genera un gráfico de regresión lineal entre las columnas 'Age' y 'OVR' del DataFrame 'original_data'.
sns.regplot(
    x='Age',                     # Variable en el eje x (Edad de los jugadores).
    y='OVR',                     # Variable en el eje y (Valoración general de los jugadores).
    data=original_data,          # El DataFrame que contiene los datos.
    color='orange',              # Color de los puntos en el gráfico (naranja).
    marker='.',                  # Estilo de los puntos en el gráfico.
    x_jitter=np.std(original_data.Age),   # Agrega un pequeño ruido aleatorio (jitter) en el eje x basado en la desviación estándar de 'Age'.
    y_jitter=np.std(original_data.OVR),   # Agrega jitter en el eje y basado en la desviación estándar de 'OVR'.
    line_kws={'alpha': 0.4, 'color': 'k'}  # Establece la transparencia (alpha) y color (negro) de la línea de regresión.
)
plt.title("Relación entre edad y Overall")
plt.xlabel("Edad")
plt.ylabel("Overall")
plt.grid(True)
plt.grid('--', alpha = 0.5)
plt.show()
```

***Top 30 de equipos según su OVR medio:***

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Agrupar por equipo y calcular la media de la calificación general
team_avg_ovr = original_data.groupby(['Team'])['OVR'].mean().sort_values(ascending=False).reset_index().head(30)

# Crear la figura para graficar los equipos clasificados según su promedio OVR
plt.figure(figsize=(8, 5))  # Ajustar el tamaño de la figura

# Graficar las barras con un paleta de colores
bar_plot = sns.barplot(x='OVR', y='Team', data=team_avg_ovr, palette='viridis', hue = 'Team')

# Aumentar el ancho de las barras
for bar in bar_plot.patches:
    bar.set_height(0.6)  # Ajustar el ancho de las barras

# Añadir título y etiquetas
plt.title('Top 30 Equipos Clasificados por Calificación General Promedio (OVR)', fontsize=16)
plt.xlabel('Calificación General Promedio (OVR)')
plt.xlim([70, 85])
plt.ylabel('Equipo')

# Añadir etiquetas de valor en cada barra
for index, (team, ovr) in enumerate(zip(team_avg_ovr['Team'], team_avg_ovr['OVR'])):
    plt.text(ovr + 0.5, index, f'{ovr:.2f}', va='center', ha='center', fontsize=10)

# Ajustar el diseño para evitar superposiciones
plt.tight_layout()

# Mostrar la gráfica
plt.show()
```

***Ranking de todas las ligas según OVR medio:***

```python
# Agrupar por liga y calcular la media de la calificación general
league_avg_ovr = original_data.groupby(['League'])['OVR'].mean().sort_values(ascending=False).reset_index()

# Crear la figura para graficar las ligas clasificadas según su promedio OVR
plt.figure(figsize=(10, 8))  # Ajustar el tamaño de la figura

# Graficar las barras con una paleta de colores
bar_plot = sns.barplot(x='OVR', y='League', data=league_avg_ovr, palette='hsv', hue = 'League')

# Aumentar el ancho de las barras
for bar in bar_plot.patches:
    bar.set_height(0.6)  # Ajustar el ancho de las barras

# Añadir título y etiquetas
plt.title('Ligas Clasificadas por Calificación General Promedio (OVR)', fontsize=16)
plt.xlabel('Calificación General Promedio (OVR)')
plt.xlim([50, 80])  # Ajustar los límites del eje X
plt.ylabel('Liga')

# Añadir etiquetas de valor en cada barra (opcional)
for index, (league, ovr) in enumerate(zip(league_avg_ovr['League'], league_avg_ovr['OVR'])):
    plt.text(ovr + 0.5, index, f'{ovr:.2f}', va='center', ha='center', fontsize=10)

# Ajustar el diseño para evitar superposiciones
plt.tight_layout()

# Mostrar la gráfica
plt.show()
```

***Mejores jugadores por cada liga:***

```python
# Agrupamos por 'League' y encontramos el índice del jugador con la calificación (OVR) más alta en cada liga.
mejores_jugadores_por_liga = original_data.loc[
    original_data.groupby('League')['OVR'].idxmax()
][['League', 'Name', 'OVR']]

# Explicación:
# - `groupby('League')`: Agrupa el DataFrame por la columna 'League' (ligas).
# - `['OVR'].idxmax()`: Dentro de cada grupo, encuentra el índice del jugador con el valor más alto en la columna 'OVR' (Overall Rating).
# - `original_data.loc[...]`: Selecciona las filas correspondientes a esos índices.
# - `[["League", "Name", "OVR"]]`: Selecciona solo las columnas relevantes: la liga ('League'), el nombre del jugador ('Name') y su calificación ('OVR').

# Restablecemos el índice del DataFrame resultante, eliminando el índice original.
mejores_jugadores_por_liga.reset_index(drop=True, inplace=True)

# Explicación:
# - `reset_index(drop=True)`: Restablece el índice de `mejores_jugadores_por_liga` eliminando el índice anterior (los índices de los jugadores seleccionados).
# - `inplace=True`: Aplica el cambio directamente al DataFrame sin necesidad de asignarlo a una nueva variable.

# Ordenamos el DataFrame por la columna 'OVR' en orden descendente y seleccionamos los 10 jugadores con las mejores calificaciones.
mejores_jugadores = mejores_jugadores_por_liga.sort_values('OVR', ascending=False).head(10).reset_index(drop=True)

# Explicación:
# - `sort_values('OVR', ascending=False)`: Ordena las filas del DataFrame por la columna 'OVR' en orden descendente (de mayor a menor calificación).
# - `head(10)`: Selecciona los primeros 10 jugadores con las calificaciones más altas.
# - `reset_index(drop=True)`: Restablece el índice nuevamente después de la ordenación y eliminación de las filas no seleccionadas.



# Mostramos el resultado final
mejores_jugadores
```
